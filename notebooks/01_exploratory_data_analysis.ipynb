#  RAVDESS

RAVDESS is actually the first dataset we encountered when searching for data for this project, it's widely used in speech-related tasks and contains recordings of 24 actors (12 female, 12 male) speaking with different emotions. The following is how their website describe the file naming convention:

"<i>
Each of the 1440 files has a unique filename. The filename consists of a 7-part numerical identifier (e.g., 03-01-06-01-02-01-12.wav). These identifiers define the stimulus characteristics:

<ul>
    <li>Modality (01 = full-AV, 02 = video-only, 03 = audio-only).</li>
    <li>Vocal channel (01 = speech, 02 = song).</li>
    <li>Emotion (01 = neutral, 02 = calm, 03 = happy, 04 = sad, 05 = angry, 06 = fearful, 07 = disgust, 08 = surprised).</li>
    <li>Emotional intensity (01 = normal, 02 = strong). NOTE: There is no strong intensity for the 'neutral' emotion.</li>
    <li>Statement (01 = "Kids are talking by the door", 02 = "Dogs are sitting by the door").</li>
    <li>Repetition (01 = 1st repetition, 02 = 2nd repetition).</li>
    <li>Actor (01 to 24. Odd numbered actors are male, even numbered actors are female).</li>
</ul>

Filename example: 03-01-06-01-02-01-12.wav

<ol>
    <li>Audio-only (03)</li>
    <li>Speech (01)</li>
    <li>Fearful (06)</li>
    <li>Normal intensity (01)</li>
    <li>Statement "dogs" (02)</li>
    <li>1st Repetition (01)</li>
    <li>12th Actor (12)</li>
</ol>
Female, as the actor ID number is even.
</i>"

## Class imbalance check

A quick look at per-class sample counts sorted from largest to smallest, plus the **imbalance ratio** (max / min) to quantify how skewed the dataset is as this will very much determine the performance of our models.


```python
# per-class counts
class_counts = df["labels"].value_counts().sort_values(ascending=True)

# derive emotion column from labels (format: "gender_emotion")
if "emotion" not in df.columns:
    df["emotion"] = df["labels"].str.split("_").str[1]

fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# left: all combined classes
colors_left = ["#E07B9B" if lbl.startswith("female") else "#5B9BD5" for lbl in class_counts.index]
class_counts.plot.barh(ax=axes[0], color=colors_left, edgecolor="black", linewidth=0.4)
axes[0].set_xlabel("Number of samples", fontsize=12)
axes[0].set_ylabel("")
axes[0].set_title("Samples per class (gender × emotion)", fontsize=13)
for i, (val, lbl) in enumerate(zip(class_counts.values, class_counts.index)):
    axes[0].text(val + 15, i, str(val), va="center", fontsize=9)
axes[0].grid(axis="x", alpha=0.3)

# right: emotion-only counts
emotion_counts = df["emotion"].value_counts().sort_values(ascending=True)
emotion_counts.plot.barh(ax=axes[1], color="#FFC857", edgecolor="black", linewidth=0.4)
axes[1].set_xlabel("Number of samples", fontsize=12)
axes[1].set_ylabel("")
axes[1].set_title("Samples per emotion (genders merged)", fontsize=13)
for i, (val, lbl) in enumerate(zip(emotion_counts.values, emotion_counts.index)):
    axes[1].text(val + 15, i, str(val), va="center", fontsize=9)
axes[1].grid(axis="x", alpha=0.3)

plt.tight_layout()
plt.show()

# imbalance summary

print(f"\n{'Class':<22} {'Count':>6}  {'% of total':>10}")
print("─" * 42)
total = len(df)
for lbl, cnt in class_counts.items():
    print(f"  {lbl:<20} {cnt:>6}  {cnt/total*100:>9.1f}%")


# emotion-level
print(f"\n{'Emotion':<15} {'Count':>6}  {'% of total':>10}")
print("─" * 35)
for emo, cnt in emotion_counts.items():
    print(f"  {emo:<13} {cnt:>6}  {cnt/total*100:>9.1f}%")
```

## PCA scatter, do emotion/gender clusters exist in feature space?

We reduce the 234 extracted features down to 2 principal components and scatter-plot every sample. If emotions or genders form visible clusters, the features are capturing meaningful signal.

```python
# load the processed feature dataset and defragment with .copy()
feat_df = pd.read_csv(DATA_PROCESSED / "dataset.csv").copy()
feat_df = feat_df.assign(
    gender=feat_df["labels"].str.split("_").str[0],
    emotion=feat_df["labels"].str.split("_").str[1],
)

meta = ["labels", "source", "path", "gender", "emotion"]
X = feat_df.drop(columns=[c for c in meta if c in feat_df.columns]).values
n_features = X.shape[1]

# standardise then reduce to 2D with PCA
X_scaled = StandardScaler().fit_transform(X)
pca = PCA(n_components=2, random_state=42)
coords = pca.fit_transform(X_scaled)
feat_df = feat_df.assign(PC1=coords[:, 0], PC2=coords[:, 1])

var1, var2 = pca.explained_variance_ratio_ * 100

emotion_list = sorted(feat_df["emotion"].unique())
emotion_cmap = plt.colormaps.get_cmap("tab10").resampled(len(emotion_list))
emotion_colors = {e: emotion_cmap(i) for i, e in enumerate(emotion_list)}

gender_colors = {"female": "#E07B9B", "male": "#5B9BD5"}
gender_markers = {"female": "o", "male": "s"}

fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# left: coloured by EMOTION
for emotion in emotion_list:
    mask = feat_df["emotion"] == emotion
    axes[0].scatter(
        feat_df.loc[mask, "PC1"], feat_df.loc[mask, "PC2"],
        c=[emotion_colors[emotion]], s=8, alpha=0.45, linewidths=0,
        label=emotion,
    )
axes[0].set_xlabel(f"PC1 ({var1:.1f}% variance)", fontsize=11)
axes[0].set_ylabel(f"PC2 ({var2:.1f}% variance)", fontsize=11)
axes[0].set_title("PCA - coloured by emotion", fontsize=13)
axes[0].legend(markerscale=2.5, fontsize=9, loc="best",
               framealpha=0.7, ncol=2)
axes[0].grid(alpha=0.2)

# right: coloured by GENDER
for gender in ["female", "male"]:
    for emotion in emotion_list:
        mask = (feat_df["gender"] == gender) & (feat_df["emotion"] == emotion)
        axes[1].scatter(
            feat_df.loc[mask, "PC1"], feat_df.loc[mask, "PC2"],
            c=[gender_colors[gender]],
            marker=gender_markers[gender],
            s=8, alpha=0.4, linewidths=0,
        )

legend_handles = [
    mpatches.Patch(color=gender_colors["female"], label="female"),
    mpatches.Patch(color=gender_colors["male"],   label="male"),
]
axes[1].legend(handles=legend_handles, fontsize=11, loc="best")
axes[1].set_xlabel(f"PC1 ({var1:.1f}% variance)", fontsize=11)
axes[1].set_ylabel(f"PC2 ({var2:.1f}% variance)", fontsize=11)
axes[1].set_title("PCA - coloured by gender", fontsize=13)
axes[1].grid(alpha=0.2)

plt.suptitle(
    f"PCA of {n_features} audio features  ({len(feat_df):,} samples)",
    fontsize=14, y=1.01,
)
plt.tight_layout()
plt.show()

print(f"Variance explained - PC1: {var1:.1f}%  PC2: {var2:.1f}%")
```

## t-SNE — non-linear dimensionality reduction

PCA is linear, so it may miss curved or twisted manifolds. **t-SNE** preserves local neighbourhood structure and is much better at revealing tight clusters if they exist. We first reduce to 30 PCs (for speed), then project to 2D with t-SNE.

```python
# speed trick: PCA to 30 dims first, then t-SNE to 2D
pca30 = PCA(n_components=30, random_state=42)
X_pca30 = pca30.fit_transform(X_scaled)
print(f"PCA 30-D captures {pca30.explained_variance_ratio_.sum()*100:.1f}% of total variance")

tsne = TSNE(n_components=2, perplexity=40, learning_rate="auto",
            init="pca", random_state=42, n_jobs=-1)
tsne_coords = tsne.fit_transform(X_pca30)
feat_df = feat_df.assign(tSNE1=tsne_coords[:, 0], tSNE2=tsne_coords[:, 1])
print("Done!")

fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# left: coloured by EMOTION
for emotion in emotion_list:
    mask = feat_df["emotion"] == emotion
    axes[0].scatter(
        feat_df.loc[mask, "tSNE1"], feat_df.loc[mask, "tSNE2"],
        c=[emotion_colors[emotion]], s=6, alpha=0.5, linewidths=0,
        label=emotion,
    )
axes[0].set_xlabel("t-SNE 1", fontsize=11)
axes[0].set_ylabel("t-SNE 2", fontsize=11)
axes[0].set_title("t-SNE - coloured by emotion", fontsize=13)
axes[0].legend(markerscale=2.5, fontsize=9, loc="best", framealpha=0.7, ncol=2)
axes[0].grid(alpha=0.2)

# right: coloured by GENDER
for gender in ["female", "male"]:
    mask = feat_df["gender"] == gender
    axes[1].scatter(
        feat_df.loc[mask, "tSNE1"], feat_df.loc[mask, "tSNE2"],
        c=[gender_colors[gender]],
        marker=gender_markers[gender],
        s=6, alpha=0.45, linewidths=0,
    )
handles = [
    mpatches.Patch(color=gender_colors["female"], label="female"),
    mpatches.Patch(color=gender_colors["male"],   label="male"),
]
axes[1].legend(handles=handles, fontsize=11, loc="best")
axes[1].set_xlabel("t-SNE 1", fontsize=11)
axes[1].set_ylabel("t-SNE 2", fontsize=11)
axes[1].set_title("t-SNE - coloured by gender", fontsize=13)
axes[1].grid(alpha=0.2)

plt.suptitle(
    f"t-SNE of {n_features} audio features  ({len(feat_df):,} samples,  perplexity=40)",
    fontsize=14, y=1.01,
)
plt.tight_layout()
plt.show()
```

some clusters are noticeable most importantly for gender...that's some good news.
